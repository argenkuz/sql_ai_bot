```python
import pandas as pd

df = pd.read_csv("dataset/spotify.csv")
df.head()
```

```python

# удаляем дубликаты
df = df.drop_duplicates()

# убираем строки без track_id
df = df.dropna(subset=['track_id'])

# чистим artists
df['artists'] = df['artists'].fillna("Unknown")
df['artists'] = df['artists'].str.split(',')
df = df.explode('artists')
df['artists'] = df['artists'].str.strip()
```

```python
artists = df[['artists']].drop_duplicates().reset_index(drop=True)
artists['artist_id'] = artists.index + 1
artists = artists.rename(columns={'artists': 'artist_name'})
```

```python
albums = df[['album_name']].drop_duplicates().reset_index(drop=True)
albums['album_id'] = albums.index + 1
```

```python
genres = df[['track_genre']].drop_duplicates().reset_index(drop=True)
genres['genre_id'] = genres.index + 1
genres = genres.rename(columns={'track_genre': 'genre_name'})
```

```python
df = df.merge(artists, left_on='artists', right_on='artist_name')
df = df.merge(albums, on='album_name')
df = df.merge(genres, left_on='track_genre', right_on='genre_name')
```

```python
tracks = df[['track_id', 'track_name', 'album_id', 'genre_id',
             'popularity', 'duration_ms', 'explicit']].drop_duplicates()

tracks = tracks.reset_index(drop=True)
```

```python
track_artist = df[['track_id', 'artist_id']].drop_duplicates()
track_artist = track_artist.reset_index(drop=True)
track_artist['id'] = track_artist.index + 1
```

```python
audio_features = df[['track_id', 'danceability', 'energy', 'key',
                     'loudness', 'mode', 'speechiness',
                     'acousticness', 'instrumentalness',
                     'liveness', 'valence', 'tempo',
                     'time_signature']].drop_duplicates()

audio_features = audio_features.reset_index(drop=True)
audio_features['feature_id'] = audio_features.index + 1
```

```python
tracks.to_csv("cleaned_dataset/tracks.csv", index=False)
artists.to_csv("cleaned_dataset/artists.csv", index=False)
albums.to_csv("cleaned_dataset/albums.csv", index=False)
genres.to_csv("cleaned_dataset/genres.csv", index=False)
track_artist.to_csv("cleaned_dataset/track_artist.csv", index=False)
audio_features.to_csv("cleaned_dataset/audio_features.csv", index=False)
```
