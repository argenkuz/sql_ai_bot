```python
import pandas as pd

df = pd.read_csv("cleaned_dataset/tracks.csv")

df.head()
```

```python
df['explicit'] = df['explicit'].map({
    True: 1,
    False: 0,
    'True': 1,
    'False': 0
})

df['explicit'].value_counts()
```

```python
duplicates = df[df.duplicated(subset=['track_id'], keep=False)]

duplicates
```

```python
df = df.drop_duplicates(subset=['track_id'])
```

```python
df.to_csv("tracks_clean.csv", index=False)
```
